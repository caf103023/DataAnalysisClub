```python
import requests
from bs4 import BeautifulSoup
res = requests.get('https://disp.cc/b/Gossiping')
soup = BeautifulSoup(res.text)
for i in soup.select('.row2'):
    print(i.select('.fgG1')[0].string, i.select('.titleColor')[0].text, i.select('.bgB')[0].text)
```
